```python
import pandas as pd
import folium
```

```python

estat2024 = pd.read_csv("../data/estat2024.csv")

e24_01_01_1200 = estat2024[(estat2024['datetime'] >= '2024-01-01 12:00:00') & (estat2024['datetime'] < '2024-01-01 12:05:00')]
```

```python
df = e24_01_01_1200


estado_colors = {
    0: "green",        # Fluid
    1: "orange",       # Dense
    2: "red",          # Congested
    3: "black",        # Closed
    4: "gray",         # No data
    5: "lightgreen",   # Underpass fluid
    6: "darkorange",   # Underpass dense
    7: "darkred",      # Underpass congestedxa
    8: "darkblue",     # Underpass closed
    9: "lightgray"     # No data (underpass)
}

df[['lat', 'lon']] = df['geo_point_2d'].str.split(",", expand=True).astype(float)

# Create folium map centered around Valencia
traffic_map = folium.Map(location=[39.4701, -0.3704], zoom_start=14)

# Add points
for _, row in df.iterrows():
    folium.CircleMarker(
        location=(row['lat'], row['lon']),
        radius=6,
        color=estado_colors.get(row['Estat / Estado'], "white"),
        fill=True,
        fill_opacity=0.9,
        popup=f"Estado: {row['Estat / Estado']}"
    ).add_to(traffic_map)

traffic_map
```
