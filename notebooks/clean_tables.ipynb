```python
def clean_valenbisi(df):
    data = df.copy()
    data['Direction'] = data['Direccion']
    data['Number'] = data['Numero']
    data['Active'] = data['Activo'].map({'T': 1, 'F': 0}).astype(int)
    data['Free_bici'] = data['Bicis_disponibles']
    data['Free_stand'] = data['Espacios_libres']
    data['Total_stands'] = data['Espacios_totales']
    data['Ticket'] = data['ticket'].map({'T': 1, 'F': 0})
    data['Update_date'] = pd.to_datetime(data['fecha_actualizacion'], format='%d/%m/%Y %H:%M:%S')
    data[['Latitude', 'Longitude']] = data['geo_point_2d'].str.split(",", expand=True).astype(float)
    data['Folder_datetime'] = data['datetime']

    data.drop(columns=['Direccion', 'Numero','Activo','Bicis_disponibles', 'Espacios_libres',
                     'Espacios_totales','ticket', 'fecha_actualizacion', 'geo_shape',
                     'geo_point_2d', 'update_jcd', 'datetime'], inplace=True)
    return data
```

```python
v24_clean = clean_valenbisi(v24)
v24_clean.head()
```

```python
min_date = v24_clean['Folder_datetime'].min()
max_date = v24_clean['Folder_datetime'].max()

print(f"Folder_datetime range: {min_date} to {max_date}")

min_date = v24_clean['Update_date'].min()
max_date = v24_clean['Update_date'].max()

print(f"Update_date range: {min_date} to {max_date}")
```

```python
v25 = pd.read_csv("data/valenbisi2025.csv")
v25_clean = clean_valenbisi(v25)

valenbisi = pd.concat([v24_clean, v25_clean], ignore_index=True)
valenbisi = valenbisi[(valenbisi['Update_date'] >= '2024-01-01 00:00:00') & (valenbisi['Update_date'] <= '2025-05-01 00:00:00')]
valenbisi.head()
```

```python
min_date = valenbisi['Update_date'].min()
max_date = valenbisi['Update_date'].max()

print(f"Update_date range: {min_date} to {max_date}")
valenbisi.info()
```

```python
valenbisi.to_csv('data/VALENBISI.csv', index=False)
```

## Get average hourly bici counts for each station

```python
import pandas as pd

def get_station_counts_df(valenbisi):
    # Filter active records only
    df = valenbisi[valenbisi['Active'] == 1].copy()
    
    # Extract hour from timestamp
    df['Hour'] = pd.to_datetime(df['Update_date']).dt.hour
    
    # Group by station number and hour, then calculate average free bikes and stands
    result = df.groupby(['Number', 'Hour']).agg(
        Avg_Free_Bici=('Free_bici', 'mean'),
        Avg_Free_Stand=('Free_stand', 'mean')
    ).reset_index()
    
    # Rename columns for clarity
    result.columns = ['StationNumber', 'HourOfDay', 'Avg_Free_Bici', 'Avg_Free_Stand']
    
    return result

# Call the function to get the DataFrame
station_counts_df = get_station_counts_df(valenbisi)

# Display the first few rows
station_counts_df.head()

# Save the result to a CSV file
station_counts_df.to_csv('streamlitdata/HourlyCount.csv', index=False)
```

```python
# add weekday column

def get_station_counts_df(valenbisi):
    # Filter active records only
    df = valenbisi[valenbisi['Active'] == 1].copy()

    # Parse timestamp
    df['Update_date'] = pd.to_datetime(df['Update_date'])

    # Extract weekday and hour
    df['Weekday'] = df['Update_date'].dt.day_name()
    df['Hour'] = df['Update_date'].dt.hour

    # Group by station, weekday, and hour to compute averages
    result = df.groupby(['Number', 'Weekday', 'Hour']).agg(
        Avg_Free_Bici=('Free_bici', 'mean'),
        Avg_Free_Stand=('Free_stand', 'mean')
    ).reset_index()

    # Optional: sort Weekday for readability
    weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    result['Weekday'] = pd.Categorical(result['Weekday'], categories=weekday_order, ordered=True)
    result = result.sort_values(by=['Number', 'Weekday', 'Hour'])

    result.columns = ['StationNumber', 'Weekday','HourOfDay', 'Avg_Free_Bici', 'Avg_Free_Stand']

    return result

# Call the function
station_weekday_df = get_station_counts_df(valenbisi)

# Preview result
print(station_weekday_df.head())

# Save to CSV
station_weekday_df.to_csv('../streamlitdata/WeekdayAverages.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# Load the data
valenbisi = pd.read_csv('data/VALENBISI.csv')

# Convert 'Update_date' to datetime if it isn't already
valenbisi['Update_date'] = pd.to_datetime(valenbisi['Update_date'])

# Sort by station number and date in descending order
valenbisi_sorted = valenbisi.sort_values(by=['Number', 'Update_date'], ascending=[True, False])

# Find the last date in the data
last_date = valenbisi_sorted['Update_date'].max()

# Filter the data for the last date
valenbisi_last_date = valenbisi_sorted[valenbisi_sorted['Update_date'].dt.date == last_date.date()]

# Function to find the row closest to 11:00 AM for each station
def get_closest_to_11am(group):
    # Calculate the difference between the update time and 11:00 AM
    eleven_am = pd.Timestamp(year=group['Update_date'].dt.year.iloc[0], 
                             month=group['Update_date'].dt.month.iloc[0],
                             day=group['Update_date'].dt.day.iloc[0], 
                             hour=11, minute=0)
    
    # Calculate the absolute time difference from 11:00 AM
    group['time_diff'] = (group['Update_date'] - eleven_am).abs()
    
    # Get the row with the minimum time difference (closest to 11:00 AM)
    closest_row = group.loc[group['time_diff'].idxmin()]
    return closest_row

# Apply the function to each station's data
closest_to_11am_per_station = valenbisi_last_date.groupby('Number').apply(get_closest_to_11am).reset_index(drop=True)

# Print the result
print(closest_to_11am_per_station.head())

# Save the result to a new CSV file
closest_to_11am_per_station.to_csv('streamlitdata/NewestData.csv', index=False)
```
